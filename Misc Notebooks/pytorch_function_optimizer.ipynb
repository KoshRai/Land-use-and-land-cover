```python
import torch
import math
import matplotlib.pyplot as plt
```

```python
def func(x):
    return x**3 + 4
```

```python
def optimize(f, x, lr):
    x = torch.tensor([x], requires_grad=True, dtype=torch.float16)
    print(f"Initially x = {x.item()}")
    optimizer = torch.optim.SGD([x], lr=lr)
    for i in range(100000):
        y = f(x)
        y.backward()
        optimizer.step()
        optimizer.zero_grad()
    print(f"Minima at {y.item()}")
    print(f"For x = {x.item()}")
```

```python
optimize(func, 1, 1)
```

```python
x = [i for i in range(-50, 50)]
y = [func(i) for i in range(-50,50)]
plt.plot(x,y)
```
